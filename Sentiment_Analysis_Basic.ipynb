Sentiment Analysis

```python
from textblob import TextBlob
text1 = "The movie was absolutely fantastic, I loved every moment of it!"
blob1 = TextBlob(text1)
sentiment1 = blob1.sentiment.polarity
print("Sentiment Analysis using TextBlob")
print("=" * 60) # print "=", 60 times
print("TEXT 1:", text1) # print as key and value
print(f"Sentiment Score: {sentiment1:.2f}")
print("Sentiment Score:", sentiment1) #2f means score till two decimal places
if 0 <sentiment1 <=0.5:
  print("Sarcastic Sentence ðŸ˜")
elif sentiment1 >= 0.3:
  print("Positive Sentence ðŸ˜Š")
else:
  print("Negative Sentence ðŸ˜’")
print("=" * 60)
```

```python
from afinn import Afinn
af = Afinn()
af1 = af.score(text1)
print("Sentiment Analysis using Afinn")
print("=" * 60)
print("Sentiment Score:", af1)
if 0 <af1 <=0.5:
  print("Sarcastic Sentence ðŸ˜")
elif af1 >= 0.3:
  print("Positive Sentence ðŸ˜Š")
else:
  print("Negative Sentence ðŸ˜’")
print("=" * 60)
```
